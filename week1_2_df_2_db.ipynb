## Pymysql을 이용하기

* Pymysql의 경우 DB 연결만 지원해주기 때문에 DB 연결을 직접 관리하고, SQL 쿼리를 수동으로 작성해서 실행해야합니다.

### 1. PyMySQL 라이브러리 import 

```python
import pymysql
```

### 2. connect 메소드를 이용해 DB에 연결.

```python
# 데이터베이스 연결 설정
connection = pymysql.connect(host='localhost',  # 데이터베이스 서버 주소
                             user='root',  # 해당 DB에서 사용하는 user 이름
                             password='0000',  # 해당 DB의 해당 user의 비밀번호
                             database='my_test',  # 접속하려는 DB명
                             charset='utf8mb4',  # 가장 기본적인 charset.
                             cursorclass=pymysql.cursors.DictCursor)
```

### 3. cursor 객체 가져오기. 

```python
cursor = connection.cursor()  # DB 내 위치.
# 위 코드 실행 후 영향을 받은 행의 수를 반환함. DDL의 경우 영향을 받은 행은 없기 때문에 0 반환. 
```

### 4. cursor의 execute() 메소드를 사용해 SQL문을 DB 서버에 보냄.

```python
# DB가 존재하지 않기 때문에 테이블 생성
create_table_query = """
CREATE TABLE IF NOT EXISTS hospital_pymysql (
    addr VARCHAR(255),
    clCdNm VARCHAR(255),
    crtrYm VARCHAR(255),
    estbCnclTp VARCHAR(255),    
    estbDd VARCHAR(255),
    telno VARCHAR(255),
    yadmNm VARCHAR(255),    
    ykiho VARCHAR(255),    
    shwSbjtCdNm VARCHAR(255)     
);
"""
cursor.execute(create_table_query)
```

### 5. DML의 경우 commit()을 사용해 데이터 갱신 확정.

```python
# DataFrame의 각 행을 데이터베이스에 삽입

insert_query = """
INSERT INTO hospital_pymysql 
VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s);
"""
# 여기서 %s는 SQL 에서 쿼리를 작성할 때 사용되는 파라미터의 플레이스 홀더. 
# 쿼리가 실행될 때 치환될 값의 위치를 나타냄.

# DataFrame에 저장된 데이터를 row-by-row(행별로) 데이터베이스에 삽입
for _, row in df.iterrows():
    cursor.execute(insert_query, (row['addr'], row['clCdNm'], row['crtrYm'], row['estbCnclTp'], 
                                  row['estbDd'], row['telno'], row['yadmNm'], row['ykiho'], row['shwSbjtCdNm']))

# 변경 사항을 데이터베이스에 반영. DML(Insert/Update/Delete)의 경우 commit을 사용해 데이터 확정 갱신. 
connection.commit()
```

실제 실행되는 쿼리 한 줄 예시:  
"INSERT INTO hospital_pymysql VALUES ("인천광역시 계양구 계양대로 85, 2층 (작전동)", "한의원", "202401", "개업", "20240103", "032-507-1075", "조각목한의원", "JDQ4MTYyMiM1MSMkMSMkMCMkMTMkNDgxMzUxIzUxIyQxIy", "");"

### 6. 데이터 가져오기

## ORM 라이브러리인 SQLAlchemy 사용

```python
import pandas as pd
from sqlalchemy import create_engine

# MySQL 연결 설정
username = 'root'
password = '0000'
host = 'localhost'  # 데이터베이스 서버 주소
port = 3306
database = 'my_test'

# pymysql을 사용하여 MySQL 연결 엔진 생성
# 연결 문자열 포맷: DB명+driver://username:password@host:port/database
engine = create_engine(f'mysql+pymysql://{username}:{password}@{host}:{port}/{database}')
```

### 1. DataFrame to DB

```python
# DataFrame을 MySQL 데이터베이스에 저장
# 'hospital_alchemy'은 저장하려는 테이블 이름. 이 테이블 이름이 데이터베이스에 없으면 새로 생성됨.
# if_exists 옵션은 테이블이 이미 존재할 경우 어떻게 할지 결정. 'fail', 'replace', 'append' 중 선택 가능.
df.to_sql('hospital_alchemy', con=engine, index=False, if_exists='replace')
```

### 2. DB to DataFrame

```python
# SQL 쿼리 (여기서는 예제로 테이블 전체를 선택)
query = "SELECT * FROM hospital_alchemy"

# SQL 쿼리 실행 및 결과를 DataFrame으로 로드
df = pd.read_sql(query, engine)

# 결과 확인
print(df)
```

## (번외) API와 반복문의 만남

```python
url = 'http://apis.data.go.kr/B551182/yadmOpCloInfoService1/getHospPharmacyOpCloList'

df = pd.DataFrame()
for my_month in ['202301', '202302', '202303']:
    params ={'serviceKey' : serviceKey, 'numOfRows' : '10', 'pageNo' : '1', 'crtrYm' : f'{my_month}', 'yadmTp' : '1', 'opCloTp' : '1' }
    print (params)
    response = requests.get(url, params=params)
    temp_df = pd.read_xml(response.content, xpath=".//item")  # 추가된 코드
    if df.empty:
        df = temp_df
    else:
        df = pd.concat([df, temp_df], ignore_index=True)
```
