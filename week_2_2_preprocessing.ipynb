# scikit-learn의 기본 메서드!

* fit: 데이터를 입력으로 받아서 모델을 학습 시킴. train set에만 사용.
* transform: fit을 통해 학습된 구조를 사용해 데이터를 변환. test set에 사용.
* fit_transform: tit과 transform을 연속적으로 수행하는 편리한 방법. train set에만 사용.
