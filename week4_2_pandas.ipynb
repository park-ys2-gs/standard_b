# lambda (익명 함수를 생성할 때 사용)

```python
def my_add(x, y):
  return x + y
print(my_add(5, 3))
```

```python
add = lambda x, y: x + y  # lambda 매개변수 : 표현식
print(add(5, 3))
```

# apply
pandas 데이터 프레임이나 시리즈에 있는 데이터에 함수를 적용할 때 사용.  
이는 행이나 열에 대해 함수를 적용하여 그 결과를 반환하게 합니다.

```python
import pandas as pd

# 데이터 프레임 생성
df = pd.DataFrame({
    'A': range(1, 5),
    'B': range(10, 50, 10)
})
print(df)
```

```python
# 모든 열의 각 값에 2를 곱하는 함수 적용
def my_multi(x):
  return x * 2
df.apply(my_multi)  # axis=0인 경우 행에 대해 연산 수행(default). 1인경우는 columns(열)에 연산 수행
```

```python
# 모든 열의 각 값에 2를 곱하는 함수 적용
df.apply(lambda x: x * 2)
```

```python
# A 열의 각 값에 2를 곱하는 함수 적용
df['A'].apply(lambda x: x * 2)
```

* apply 함수의 axis 매개변수는 함수가 적용되는 방향을 결정합.
* axis=0은 함수를 각 열에 대해 수직으로 적용 (열 연산).
* axis=1은 함수를 각 행에 대해 수평으로 적용 (행 연산).

```python
# 데이터 프레임 생성
df = pd.DataFrame({
    'A': range(1, 5),
    'B': range(10, 50, 10)
})
print (df)

# 각 열의 최대값과 최소값의 차이를 계산하는 함수
def range_diff(x):
    return x.max() - x.min()

# 열 연산: 각 열의 최대값과 최소값의 차이 계산
column_diff = df.apply(range_diff, axis=0)
print("열 연산 결과:")
print(column_diff)

# 행 연산: 각 행의 최대값과 최소값의 차이 계산
row_diff = df.apply(range_diff, axis=1)
print("행 연산 결과:")
print(row_diff)
```

# groupby

pandas에서 데이터를 특정 기준에 따라 그룹으로 묶고, 각 그룹에 대해 집계, 변환, 필터링 같은 연산을 쉽게 적용할 수 있도록 함.

```python
# 데이터 프레임 생성
df = pd.DataFrame({
    'Key': ['A', 'B', 'A', 'B'],
    'Data': [10, 20, 30, 40]
})
df
```

```python
# Key 열을 기준으로 그룹화하고, 각 그룹의 데이터 합계 계산
grouped = df.groupby('Key')
grouped.sum() # mean, median, min, max 등 다양한 수식 적용 가능.
```

# Pivot
데이터 프레임의 데이터를 재구조화. 이를 통해 데이터를 요약하고 분석하는 데 유용하게 사용됨.

```python
# 데이터 프레임 생성
df = pd.DataFrame({
    'Date': ['2021-01-01', '2021-01-01', '2021-01-02', '2021-01-02'],
    'Type': ['A', 'B', 'A', 'B'],
    'Value': [100, 200, 300, 400]
})
df
```

```python
# Date와 Type을 인덱스 및 컬럼으로 사용하여 피벗
pivot_df = df.pivot(index='Date', columns='Type', values='Value')
pivot_df
```
